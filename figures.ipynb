```python
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# Set the default color cycle
plt.style.use('default')
mpl.rcParams['axes.prop_cycle'] = mpl.cycler(color=["#f27970", "#32b897", "6f6f6f", "#f8ac8c","#9ac9db","2878b5"]) 


gs = gridspec.GridSpec(2, 2)


plt.figure(figsize=(10, 5),dpi=300)
x = np.array([0, 1.5, 3])
patterns = [ "xx" , "\\\\" , ".." , "--" , "++" , ""]

ax1 = plt.subplot(gs[0, :])
f1 = [1, 0.43, 0.6]
f2 = [1,0.41,0.58]
f3 = [0.24,0.76,0.37]
f4=[0.12,0.99,0.21]
f5=[0.64,0.72,0.67]
f6 = [1,0.71,0.83]
ax1.bar(x-0.5, f1, width=0.2,edgecolor='black',hatch=patterns[0])
ax1.bar(x-0.3, f2, width=0.2,edgecolor='black',hatch=patterns[1])
ax1.bar(x-0.1, f3, width=0.2,edgecolor='black',hatch=patterns[2])
ax1.bar(x+0.1, f4, width=0.2,edgecolor='black',hatch=patterns[3])
ax1.bar(x+0.3, f5, width=0.2,edgecolor='black',hatch=patterns[4])
ax1.bar(x+0.5, f6, width=0.2,edgecolor='black',hatch=patterns[5])
ax1.set_xticks([0,1.5,3])
ax1.set_yticks([0,0.5,1])
ax1.set_xticklabels(["Precision","Recall","F-measure"],fontsize=18)
ax1.set_yticklabels([0,0.5,1],fontsize=16)

ax1.set_ylabel('Score', fontsize=20)
ax1.set_xlabel('HDFS', fontsize=20)


ax2 = plt.subplot(gs[1, :])
p1 = [0.35, 0.01, 0.01]
p2 = [0.35,0.01,0.01]
p3 =  [0.14,1,0.25]
p4=[0.56,0.98,0.71]
p5=[0.53,0.67,0.33]
p6 = [0.95,0.98,0.97]
ax2.bar(x-0.5, p1, width=0.2,edgecolor='black',hatch=patterns[0])
ax2.bar(x-0.3, p2, width=0.2,edgecolor='black',hatch=patterns[1])
ax2.bar(x-0.1, p3, width=0.2,edgecolor='black',hatch=patterns[2])
ax2.bar(x+0.1, p4, width=0.2,edgecolor='black',hatch=patterns[3])
ax2.bar(x+0.3, p5, width=0.2,edgecolor='black',hatch=patterns[4])
ax2.bar(x+0.5, p6, width=0.2,edgecolor='black',hatch=patterns[5])
ax2.set_xticks([0,1.5,3])
ax2.set_yticks([0,0.5,1])
ax2.set_xticklabels(["Precision","Recall","F-measure"],fontsize=18)
ax2.set_yticklabels([0,0.5,1],fontsize=16)

ax2.set_ylabel('Score', fontsize=20)
ax2.set_xlabel('Lustre', fontsize=20)
plt.tight_layout()

plt.legend(["Decision Tree","SVM","DeepLog","LogAnomaly","NeuralLog","Drill"],loc='upper center', bbox_to_anchor=(0.5, 2.8), ncol=6, 
          prop={'size': 12})
plt.show()
# plt.savefig('diff_datasets.png', dpi=300, bbox_inches='tight')
```
